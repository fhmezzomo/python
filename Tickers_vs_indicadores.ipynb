```python
compare = ['BTC-USD', '^IXIC', '^GSPC']
data_inicial = '2016-4-1'
data_final = '2023-9-20'
compare = pdr.get_data_yahoo(compare, data_inicial, data_final)['Adj Close']
# remover finais de semana
compare = compare.dropna()
compare=compare.rename(columns={'^GSPC': 'SP500', '^IXIC': 'NASDAQ'})
compare
```

```python
compare.iloc[0]
(compare/compare.iloc[0] * 100).plot(figsize = (22,6));

# adicionar linhas verticais nas datas dos halvings
from datetime import datetime
import pandas as pd
# using a pandas timestamp:
plt.axvline(pd.Timestamp("2020-05-11"), color='g', lw=1)
# using a datetime object:
plt.axvline(datetime(2016,7,9), color='g', lw=1)
plt.text(17017, 15000, 'Verticais = halvings', size=20, color='g')
plt.title('BTC vs NASDAQ e SP500 normalizado', fontsize=20)
plt.grid(True)
plt.show
```
